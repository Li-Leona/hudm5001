```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
```

(a) Read the data into Python

```python
college = pd.read_csv("College.csv")
```

(b) Clean the data by setting the university names as the index

```python
# View the initial data (optional)
college.head()
```

```python
# Load with index set to the first column
college2 = pd.read_csv("College.csv", index_col=0)

# Rename and set index properly
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')

# Set final version of the dataset
college = college3
```

```python
college3.head()
```

(c) Numerical summary of the dataset

```python
college.describe()
```

(d) Scatterplot matrix of Top10perc, Apps, and Enroll

```python
scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']], figsize=(10, 8), alpha=0.7)
plt.suptitle('Scatterplot Matrix of Top10perc, Apps, and Enroll')
plt.show()
```

(e) Side-by-side boxplots of Outstate vs Private

```python
sns.boxplot(x='Private', y='Outstate', data=college)
plt.title('Out-of-State Tuition by Private/Public')
plt.show()
```

(f) Create a new variable Elite based on Top10perc > 50%

```python
# Create 'Elite' column
college['Elite'] = pd.cut(college['Top10perc'],
                          bins=[-1, 50, 100],  # Notice -1 instead of 0 to include 0
                          labels=['No', 'Yes'])

# Count Elite vs Non-Elite colleges
elite_counts = college['Elite'].value_counts()
print(elite_counts)

# Boxplot: Outstate vs Elite
sns.boxplot(x='Elite', y='Outstate', data=college)
plt.title('Out-of-State Tuition by Elite Status')
plt.show()
```

(g) Histograms of a few quantitative variables

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
college['Apps'].plot.hist(bins=20, ax=axes[0, 0], color='skyblue', edgecolor='black', title='Applications')
college['Accept'].plot.hist(bins=20, ax=axes[0, 1], color='lightgreen', edgecolor='black', title='Acceptances')
college['Enroll'].plot.hist(bins=20, ax=axes[1, 0], color='salmon', edgecolor='black', title='Enrollments')
college['Grad.Rate'].plot.hist(bins=20, ax=axes[1, 1], color='orange', edgecolor='black', title='Graduation Rate')

plt.tight_layout()
plt.show()
```

(h) Brief Data Exploration Summary

Here’s a brief summary based on the plots and data:

Private vs Public Colleges: Private institutions generally have significantly higher out-of-state tuition.

Elite Colleges: Elite colleges (those with more than 50% of students from the top 10% of their high school class) tend to have higher tuition and may indicate better selectivity or reputation.

Distributions:

Applications, acceptances, and enrollments are right-skewed, meaning a few universities receive vastly more applications than most.

Graduation rate is fairly normally distributed, but some colleges have 100% graduation rate, which could be worth investigating.

Scatterplots show that:

More applications loosely correlate with more enrolled students.

A higher percentage of students from top 10% correlates with more applications, hinting at reputation effects.
